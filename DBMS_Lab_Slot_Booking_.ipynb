```python
import pandas as pd

# Load the Excel file
data = pd.read_excel('lab_bookings_sample.xlsx')
data.head()
```

```python
import sqlite3

# Create an in-memory SQLite database
conn = sqlite3.connect(':memory:')
cur = conn.cursor()

# Create table
cur.execute("""
CREATE TABLE booked_slots (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    section TEXT,
    slot TEXT,
    subject TEXT,
    lab_slot TEXT,
    faculty_name TEXT,
    lab_venue TEXT
)
""")
```

```python
# Insert data into the database
for index, row in data.iterrows():
    cur.execute("""
    INSERT INTO booked_slots (section, slot, subject, lab_slot, faculty_name, lab_venue)
    VALUES (?, ?, ?, ?, ?, ?)
    """, (row['Section'], row['Slot'], row['Subject'], row['Lab Slot'], row['Faculty Name - Main 1'], row['Lab Venue']))

conn.commit()
```

```python
# Query all bookings
cur.execute("SELECT * FROM booked_slots")
bookings = cur.fetchall()

# Display bookings
import pandas as pd
bookings_df = pd.DataFrame(bookings, columns=['ID', 'Section', 'Slot', 'Subject', 'Lab Slot', 'Faculty Name', 'Lab Venue'])
bookings_df
```

```python
# Save the DataFrame to an Excel file
bookings_df.to_excel('updated_bookings.xlsx', index=False)
files.download('updated_bookings.xlsx')
```
